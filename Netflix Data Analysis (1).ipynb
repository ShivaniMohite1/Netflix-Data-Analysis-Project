```python
# What is number of Movies and TV Shows added per month?

df["month-date_added"]=df["date_added"].dt.month_name()
df["month-date_added"].value_counts()
```

```python
%matplotlib inline
plt.figure(figsize=(8,6))
x=df["month-date_added"].value_counts().plot(kind="bar",color=sns.color_palette("ch:s=.25,rot=-.25"),edgecolor="black")
for i in x.patches:
    x.annotate(int(i.get_height()),(i.get_x()+0.25,i.get_height()+1),ha="center",va="bottom")
    x.tick_params(axis="both",labelsize=10)
plt.title("show added per month\n",fontsize=15)
```

```python
# What is number of Movies and TV Shows added per year?

df["year-date_added"]=df["date_added"].dt.year
df["year-date_added"].value_counts()
```

```python
%matplotlib inline
plt.figure(figsize=(8,6))
x=df["year-date_added"].value_counts().plot(kind="bar",color=sns.color_palette("ch:s=.25,rot=-.25"),edgecolor="black")
for i in x.patches:
    x.annotate(int(i.get_height()),(i.get_x()+0.25,i.get_height()+1),ha="center",va="bottom")
    x.tick_params(axis="both",labelsize=10)
plt.title("shows added per year\n",fontsize=15)
```
